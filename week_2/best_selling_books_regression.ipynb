```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

# read dataset and transform to dataframe
df = pd.read_csv("/content/drive/MyDrive/sample_data/bestsellers with categories.csv")
df.head()
```

```python
# get mean, min, and max from dataframe
df.describe()
```

```python
df.columns
```

```python
features = ['User Rating','Reviews']
# specify prediction target
y = df.Price
# select features
X = df[features]
```

```python
# normalise dataframe
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns = X.columns)
scaled_data.head()
```

```python
# divide dataframe for testing and training
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
# make logistic regression
# train dataframe
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
